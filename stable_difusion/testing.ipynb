Gray Scale Conversion

```python
from PIL import Image
import os

def convert_depth_maps_to_single_channel(input_folder, output_folder):
    # Create output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # List all PNG files in the input folder
    for file_name in os.listdir(input_folder):
        if file_name.endswith('.png'):
            file_path = os.path.join(input_folder, file_name)
            
            # Open the image
            image = Image.open(file_path)
            
            # Convert the image to grayscale
            image = image.convert("L")
            
            # Save the single-channel image
            output_file_path = os.path.join(output_folder, file_name)
            image.save(output_file_path, "PNG")

# Example usage
input_folder = 'monocular_dataset/li/data_256/test/depth'  # Change this to your input folder path
output_folder = 'monocular_dataset/li/data_256/test/depth'  # Change this to your output folder path
convert_depth_maps_to_single_channel(input_folder, output_folder)
```
